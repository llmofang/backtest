```python
import pandas as pd
```

```python
df=pd.read_csv('stock_2016-02-26_tick1_processed.csv')
```

```python
dfs=df.groupby(['sym'])
```

```python
datas={}
for name,group in df.groupby('sym'):
    datas[name]=group
```

```python
for (k, v)  in datas.items():
    k=k[1:-1]
    v.drop_duplicates()
    v.to_csv('stock_'+k+'_2016-02-26_tick.csv')
```
